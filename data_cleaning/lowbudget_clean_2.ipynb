```python
#Import needed modules

import pandas as pd
from pandas import DataFrame, read_csv
from bs4 import BeautifulSoup as soup
from pretty_html_table import build_table
```

```python
#Load clean csv to manipulate

file_to_load = "..\data_cleaning\export\movie_db.csv"
lowest_budget = pd.read_csv(file_to_load)
lowest_budget.head(5)
```

```python
#return columns needed for web page build

lowest_total_budget = lowest_budget.loc[:, ["title", "budget", "popularity"]]
lowest_total_budget.head(5)
#lowest_total_budget.columns = ["Title", "Budget", "Populartiy"]
```

```python
#Creating DataFrame for "Total Lowest Budget Movies"

lowest_budget_sorted = lowest_total_budget.sort_values("budget").nsmallest(25, "budget")
#lowest_budget_sorted2 = lowest_budget_sorted.drop(['budget_bins'], axis=1)
print(lowest_budget_sorted)
```

```python
#Creating DataFrame for Most Popular Low Budget Movies"


popular_movie = lowest_total_budget[(lowest_total_budget["budget"] <= 1000000)]
clean_popular_movie = popular_movie.sort_values("popularity", ascending=False)
print(clean_popular_movie)
```

```python
#Creating DataFrame for least Popular Low Budget Movies

popular_movie = lowest_total_budget[(lowest_total_budget["budget"] <= 1000000)]
clean_least_popular = popular_movie.sort_values("popularity")
print(clean_least_popular)
```
